## Analise de Gama

Notebook criado para verificar se há conversão gama

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def calcular_gama(imagem):
    """
    Esta função calcula o valor gama de uma imagem
    Parâmetros:
    imagem (numpy.ndarray): A imagem a ser analisada.
    
    Retorna:
    float: Retorna o valor gama calculado.
    """
    # Normalizar a imagem para o intervalo [0, 1]
    imagem_normalizada = imagem.astype(np.float32) / 255.0
    
    # Calcular o valor gama da imagem
    gama_calculado = np.mean(np.log(imagem_normalizada + 1e-8) / np.log(0.5))
    
    # Comparar o valor gama calculado com o valor gama esperado
    return gama_calculado

```

```python
def plotar_linha_imagem(imagem, indice_linha):
    """
    Esta função recebe uma imagem e um índice de linha, e plota todos os valores dos pixels nessa linha.
    
    Parâmetros:
    imagem (numpy.ndarray): A imagem a ser analisada.
    indice_linha (int): O índice da linha a ser plotada.
    
    Retorna:
    None
    """
    # Verificar se o índice da linha está dentro do intervalo válido
    if indice_linha < 0 or indice_linha >= imagem.shape[0]:
        raise ValueError("Índice da linha fora do intervalo válido.")
    
    # Obter os valores dos pixels na linha especificada
    linha_pixels = imagem[indice_linha, :]
    
    # Plotar os valores dos pixels
    plt.figure(figsize=(10, 5))
    plt.plot(linha_pixels, marker='o')
    plt.title(f'Valores dos Pixels na Linha {indice_linha}')
    plt.xlabel('Índice do Pixel')
    plt.ylabel('Valor do Pixel')
    plt.grid(True)
    plt.show()
```
